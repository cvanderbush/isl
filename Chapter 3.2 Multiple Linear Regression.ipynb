```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
sm.show_versions()
```

```python
data = pd.read_csv('Advertising.csv')
```

```python
data.columns
```

```python
data=data.drop('Unnamed: 0', axis=1)
data
```

### Fit Ordinary Least Squares model

```python
results_radio = smf.ols('sales ~ radio', data=data).fit()
```

```python
results_radio.summary()
```

```python
results_newsp = smf.ols('sales ~ newspaper', data=data).fit()
```

```python
results_newsp.summary()
```

```python
rad_ypoints
```

```python
plt.figure(figsize = (12,8))
plt.scatter(data.radio, data.sales)
plt.xlabel('Radio spending', fontsize=20)
plt.ylabel('Sales', fontsize=20)
#construct line of best fit
rad_xpoints = pd.DataFrame({'radio':[data.radio.min(), data.radio.max()]}) # - domain of the data - this must be a
                         # a DataFrame with relevant variable as column name 
rad_ypoints = results_radio.predict(rad_xpoints) # predicted ys for the domain pts as DataFrame
plt.plot(rad_xpoints, rad_ypoints, 'r')  # plot line based on min and max x and predicted ys
```

```python
plt.figure(figsize = (12,8))
plt.scatter(data.newspaper, data.sales)
plt.xlabel('Newspaper spending', fontsize=20)
plt.ylabel('Sales', fontsize=20)
# construct the line of best fit
news_xpoints = pd.DataFrame({'newspaper':(data.newspaper.min(), data.newspaper.max())}) # domain of the data - this must be a
                         # a DataFrame with relevant variable as column name 
news_ypoints = results_newsp.predict(news_xpoints)
plt.plot(news_xpoints, news_ypoints, 'r')  # plot line based on min and max x and corresponding predicted ys
```
